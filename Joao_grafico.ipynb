```python
import numpy as np
import matplotlib.pyplot as plt

# Criando um intervalo de valores de -2 a 2
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Calculando seno e cosseno
seno = np.sin(x)
cosseno = np.cos(x)

# Criando a figura e os grficos
plt.figure(figsize=(8, 5))
plt.plot(x, seno, label='Seno', linestyle='-', linewidth=2)
plt.plot(x, cosseno, label='Cosseno', linestyle='--', linewidth=2)

# Adicionando rtulos e ttulo
plt.xlabel('ngulo (radianos)')
plt.ylabel('Valor')
plt.title('Grficos de Seno e Cosseno')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()

# Mostrando o grfico
plt.show()
```
